```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# KNN Algoritması

```python
data=pd.read_csv("data.csv")
```

```python
data.drop(["Unnamed: 32","id"],axis=1,inplace=True)
data.head()
```

```python
# malignant-->M  kötü huylu tümor
# benign-->B iyi huylu tümor
m=data[data["diagnosis"]=="M"]
b=data[data["diagnosis"]=="B"]
```

```python
plt.figure(figsize=(8,8))
plt.scatter(m["radius_mean"],m["texture_mean"],color="r",label="malignant",alpha=0.5)
plt.scatter(b["radius_mean"],b["texture_mean"],color="b",label="benign")
plt.xlabel("radius mean")
plt.ylabel("texture mean")
plt.legend()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(m["perimeter_mean"],m["area_mean"],color="r",label="M")
plt.scatter(b["perimeter_mean"],b["area_mean"],color="b",label="B")
plt.xlabel("Tümörün ortalama çevresi")
plt.ylabel("Tümörün ortalama alanı")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(m["concavity_mean"],m["compactness_mean"],color="r",label="M")
plt.scatter(b["concavity_mean"],b["compactness_mean"],color="b",label="B")
plt.xlabel("Yoğunluğun ortalama değeri")
plt.ylabel("Çukurlukların ortalama değeri")
plt.legend()
plt.show()
```

```python
data["diagnosis"]=[1 if i=="M" else 0 for i in data["diagnosis"]]
x=data.drop(["diagnosis"],axis=1).values
y=data["diagnosis"].values
#normalization
X=(x-np.min(x))/(np.max(x)-np.min(x))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=48)
```

```python
from sklearn.neighbors import KNeighborsClassifier

for i in range(1,20):
  model=KNeighborsClassifier(n_neighbors=i)
  model.fit(X_train,y_train)
  model.predict(X_test)
  print(f"{i} tane knn'li model skoru: {model.score(X_test,y_test)}")
```
